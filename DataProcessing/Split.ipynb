# **Split Data for Modeling**

Import and Load Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tabulate import tabulate
import os
```

Mount Google Drive

```python
from google.colab import drive
drive.mount('/content/drive')
```

Import Files

```python
# Define the correct folder path
base_path = "/content/drive/MyDrive/Capstone Project/Data/"

# Load feature datasets
data = pd.read_csv(base_path + "cleaned_data.csv")
```

Train-Test Split

```python
X = data.drop(columns=["Total Charges"])
y = data["Total Charges"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("Split Summary")
print(tabulate([
    ["Features Train", f"{X_train.shape[0]:,}", f"{X_train.shape[1]:,}"],
    ["Features Test", f"{X_test.shape[0]:,}", f"{X_test.shape[1]:,}"],
    ["Cost Train", f"{y_train.shape[0]:,}", y.name],
    ["Cost Test", f"{y_test.shape[0]:,}", y.name]],
    headers=["Dataset", "Rows", "Features"], tablefmt="rounded_outline"))
```

Save Datasets

```python
# Save split data
train_test = {"features_train.csv": X_train,
            "features_test.csv": X_test,
            "cost_train.csv": y_train,
            "cost_test.csv": y_test}

for file_name, mapping in train_test.items():
    full_path = os.path.join(base_path, file_name)
    with open(full_path, "w") as f:
        mapping.to_csv(full_path, index=False)
    print(f"File saved to {full_path}")
```
